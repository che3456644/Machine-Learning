```python
X=df.drop('price',axis='columns')
X
```

```python
Y=df.price
Y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
```

```python
x=4500
landareainSqFt=[[x]]
predictmodelresult = model.predict(landareainSqFt)
print(predictmodelresult)
```

```python
m=model.coef_
print(m)
```

```python
b=model.intercept_
print(b)
```

```python
y=m*x+b
print("price of {0} square feet land is : {1}".format(x,y[0]))
```
