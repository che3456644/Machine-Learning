# **Multiple Linear Regression - Economics Dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset = pd.read_csv("economic_index.csv")
```

```python
dataset.head()
```

```python
## drop unneccessary columns
dataset.drop(columns=["Unnamed: 0","year","month"],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
## check null values
dataset.isnull().sum()
```

```python
## some visualization
import seaborn as sns
sns.pairplot(dataset)
```

```python
dataset.corr()
```

```python
## visualize the datapoints more closely
plt.scatter(dataset['interest_rate'],dataset['unemployment_rate'],color='black',marker='*')
plt.xlabel("interest rate")
plt.ylabel("unemployment rate")
```

```python
## independent and dependent features
X=dataset.iloc[:,:-1]      ##    <---- OR ---->   [["interest_rate","unemployment_rate"]]
y=dataset.iloc[:,-1]       ##    <---- OR ---->   ['index_price']
```

```python
X.head()
```

```python
y
```

```python
## train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
import seaborn as sns
sns.regplot(dataset, x='interest_rate', y='index_price')
```

```python
sns.regplot(dataset, x='interest_rate', y='unemployment_rate')
```

```python
sns.regplot(dataset, x='index_price', y='unemployment_rate')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
validation_score=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=3)
```

```python
np.mean(validation_score)
```

```python
# prediction
y_pred=regression.predict(X_test)
```

```python
y_pred
```

```python
## performance metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print(score)
## display adjusted R-squared
print(1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
```
